```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('../data/raw/features_30_sec.csv')
print(df.shape)
df.head()
```

```python
print(df.columns.tolist())
```

```python
print(df['label'].value_counts())

df['label'].value_counts().plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Number of Samples per Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../outputs/figures/class_distribution.png')
plt.show()
```

```python
print("Missing values:", df.isnull().sum().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())

# Also check duplicate filenames specifically
print("Duplicate filenames:", df['filename'].duplicated().sum())
```

```python
X = df.drop(columns=['filename', 'label'])
y = df['label']

print("Feature matrix shape:", X.shape)
print("Labels shape:", y.shape)
```

```python
print(X.columns.tolist())
```

```python
X.describe()
```

```python
features_to_plot = ['tempo', 'spectral_centroid_mean', 'zero_crossing_rate_mean', 'mfcc1_mean']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for ax, feature in zip(axes.flatten(), features_to_plot):
    for genre in y.unique():
        subset = X[feature][y == genre]
        ax.hist(subset, alpha=0.5, label=genre, bins=20)
    ax.set_title(feature)
    ax.set_xlabel('Value')

axes[0][0].legend(fontsize=6)
plt.tight_layout()
plt.savefig('../outputs/figures/feature_distributions.png')
plt.show()
```
